```python
y = StockDataset1['Adj_Close']
```

```python
plt.plot(x,y, 'r') # 'r' is the color red
plt.xlabel('DATE')
plt.ylabel('PRICE')
plt.title('10 year graph')
```

```python
sma5 = talib.SMA(y,timeperiod=5)
```

```python
sma20 = talib.SMA(y,timeperiod=20)
```
